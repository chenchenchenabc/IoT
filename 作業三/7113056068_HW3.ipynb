HW3.1

***Prompt***:

1. 產生300個0~1000的亂數並依照數字大小做排序 
2. y=1時x介於500~800，其餘的y=0
3. 同時輸出兩張圖,兩張圖都有true label,但一張以logistic regression 輸出預測結果(y1)並畫上決定這些預測點與logistic regression Probability ,另一張以SVM輸出預測結果(y2)並畫上決定這些預測點與SVM Probability

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score

# 1. 產生300個0~1000的亂數並依照數字大小做排序
np.random.seed(42)
data = np.random.randint(0, 1001, size=300)
sorted_data = np.sort(data)

# 2. y=1時x介於500~800，其餘的y=0
labels = np.where((sorted_data >= 500) & (sorted_data <= 800), 1, 0)

# 將數據重塑為2D數組
X = sorted_data.reshape(-1, 1)

# 3. 建立邏輯回歸模型
log_reg = LogisticRegression()
log_reg.fit(X, labels)
y_pred_log_reg = log_reg.predict(X)
proba_log_reg = log_reg.predict_proba(X)[:, 1]

# 3. 建立支持向量機模型
svm_model = svm.SVC(probability=True)
svm_model.fit(X, labels)
y_pred_svm = svm_model.predict(X)
proba_svm = svm_model.predict_proba(X)[:, 1]

# 繪製邏輯回歸的圖
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X, labels, label='True Labels', color='blue', alpha=0.5)
plt.scatter(X, y_pred_log_reg, label='Predictions (Logistic Regression)', color='orange', alpha=0.5)
plt.plot(X, proba_log_reg, label='Probability (Logistic Regression)', color='red')
plt.title('Logistic Regression Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.axhline(0.5, color='grey', linestyle='--')
plt.legend()

# 繪製支持向量機的圖
plt.subplot(1, 2, 2)
plt.scatter(X, labels, label='True Labels', color='blue', alpha=0.5)
plt.scatter(X, y_pred_svm, label='Predictions (SVM)', color='orange', alpha=0.5)
plt.plot(X, proba_svm, label='Probability (SVM)', color='red')
plt.title('SVM Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.axhline(0.5, color='grey', linestyle='--')
plt.legend()

plt.tight_layout()
plt.show()
```
