請生成一段 Python 程式碼，使用 Hugging Face 的 `diffusers` 庫，載入 `stable-diffusion-2-1-base` 模型，並使用 `torch` 進行推理，生成一張基於提示語 "a happy dog surfing on the lake" 的圖像。程式碼應該包含以下步驟：
1. 安裝並導入所需的庫（`diffusers`, `torch`, `matplotlib`）。
2. 載入並初始化 Stable Diffusion 模型，並確保使用 GPU。
3. 設定提示語 "a happy dog surfing on the lake"。
4. 生成圖像並顯示出來。


```python
!pip install diffusers transformers torch

from diffusers import StableDiffusionPipeline
import torch

# 加載 Stable Diffusion 模型
model_id = "stabilityai/stable-diffusion-2-1-base"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")  # 確保使用 GPU

# 設定提示語
prompt = "a happy dog surfing on the lake"

# 生成圖片
image = pipe(prompt).images[0]

# 顯示圖片
import matplotlib.pyplot as plt
plt.imshow(image)
plt.axis('off')  # 關閉坐標軸
plt.show()
```
